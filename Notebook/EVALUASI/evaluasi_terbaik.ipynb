```python
# === 1. Import Library ===
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# === 2. Load Data Evaluasi Akhir ===
data_path = r"C:\Users\Lenovo\Klasifikasi_ikan\hasil_model\final_model\final_evaluation_data.npz"
data = np.load(data_path, allow_pickle=True)

X_val_final = data["X_val_final"]
y_val_final = data["y_val_final"]
y_pred_final = data["y_pred_final"]

print("✅ Data evaluasi akhir berhasil dimuat!")
print(f"Jumlah data validasi akhir: {len(X_val_final)}")

# === 3. Konversi ke label (karena one-hot encoding) ===
y_true = np.argmax(y_val_final, axis=1)
y_pred = np.argmax(y_pred_final, axis=1)

# === 4. Confusion Matrix ===
cm = confusion_matrix(y_true, y_pred)
labels = ["Segar", "Tidak Segar"]

print("\n=== Confusion Matrix ===")
print(cm)

# Tampilkan Confusion Matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - Final Model")
plt.show()

# === 5. Classification Report ===
print("\n=== Classification Report ===")
report = classification_report(y_true, y_pred, target_names=labels, digits=4)
print(report)

# === 6. Simpan hasil evaluasi ke file TXT ===
eval_path = r"C:\Users\Lenovo\Klasifikasi_ikan\hasil_model\final_model\final_evaluation_report.txt"
with open(eval_path, "w") as f:
    f.write("=== Confusion Matrix ===\n")
    f.write(str(cm))
    f.write("\n\n=== Classification Report ===\n")
    f.write(report)

print(f"\n✅ Hasil evaluasi akhir disimpan di:\n{eval_path}")
```
